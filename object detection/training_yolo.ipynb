```python
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt
import yaml

# 1. Load model
model = YOLO('yolov8m.pt')  # Use 'yolov8s.pt' etc. for larger models

# 2. Train on coco128 dataset
# Ultralytics auto-downloads the coco128.yaml and data
results = model.train(
    data='coco128.yaml',    # dataset
    epochs=30,
    imgsz=640,
    batch=16,
    project='runs',
    name='yolo_coco128_python',
    verbose=True
)

# 3. Evaluate on validation set
metrics = model.val()

# 4. Load results for plotting
# metrics dictionary: metrics.box.p, metrics.box.r, metrics.box.map50, etc.
print("Metrics summary:")
print(metrics)

# 5. Plot training results
# It automatically logs to 'runs/detect/yolo_coco128_python/results.csv'

```

### 📐 mAP Metrics Analysis (Epoch-wise)

#### 🔢 Metrics Definitions

| Metric            | Description                                                                         |
| ----------------- | ----------------------------------------------------------------------------------- |
| **Precision**     | Percentage of correct positive predictions out of all positive predictions.         |

| **mAP\@0.5**      | Mean Average Precision at IoU threshold 0.5 — emphasizes correct class and box.     |
| **mAP\@0.5:0.95** | Average mAP across IoU thresholds from 0.5 to 0.95 — stricter and more informative. |
